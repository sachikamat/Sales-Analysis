## DATA PREPROCESSING

**Find duplicate values**

```
df[df.duplicated()]
```

```
# Now remove the duplicated values.
df.drop_duplicates(inplace=True)
```

```
# Now, checking again for duplicate values.
df.duplicated().sum()
```

```
# Find duplicate values.
df.duplicated().sum()
```

**Find if the dataset contains any null values**

```
df.isnull().sum()
```

**Check empty columns**

```
df.info()
```

From the above data, it is clear that the 'Status' and 'unnamed' columns are empty, thus I'll delete them.

```
df.drop(['Status', 'unnamed1'], axis=1, inplace=True)
```

Again, checking the dataset.

```
df.info()
```

Since the empty columns are removed, now we'll check for null values.

```
df.isnull().sum()
```

```
df[df.isnull().any(axis=1)]
```

Since 12 records have null values, they should be dropped.

```
df.dropna(inplace=True)
```

```
df.isnull().sum()
```

```
df.info()
```

```
# to obtain mathematical description of different columns
df[['Age', 'Orders', 'Amount']].describe()
```

## Exploratory Data Analysis

```
df.columns
```

**Gender distribution of consumers**

```
ax = sns.countplot(x='Gender', data=df)

# To show values as well
for bars in ax.containers: 
    ax.bar_label(bars)
ax.set_ylabel('Number of consumers')
ax.set_title('Gender distribution')
```

**Finding the amount of sales done by each gender group**

```
s = df.groupby('Gender').sum()['Amount'].sort_values()
# To display data in proper tabulated format
display(HTML(s.to_frame().to_html()))
```

```
# Visualizing the amount of sales done by each gender.
plt.bar(s.index, s.values, color = ['tab:cyan', 'tab:purple'])
plt.xlabel('Gender')
plt.ylabel('Amount')
plt.title('Amount of sales done by each gender')
```

**Above visualization reveals that most consumers during the Diwali sales were female and their purchase was also greater than men.**

### Age

```
ax = sns.countplot(x='Age Group', data=df, hue = 'Gender')
for bars in ax.containers:
    ax.bar_label(bars)
ax.set_ylabel('Number of consumers')
ax.set_title('Age distribution')
```

```
# Distribution of sales amount as per age group
sales_age = df.groupby('Age Group').sum()['Amount']
sns.barplot(x = sales_age.index, y = sales_age.values)
```

**From above graph, it is clear that the individuals in the 26-35 age group contributed to the highest number of sales**
