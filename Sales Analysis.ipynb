# SALES DATA ANALYSIS


## IMPORT LIBRARIES


```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from IPython.display import display, HTML
```

## ACCESS DATA

```
df = pd.read_csv('Diwali Sales Data.csv')
```

## UNDERSTANDING THE DATASET

```
df.head()
```

## DATA PREPROCESSING

**Find duplicate values**

```
df[df.duplicated()]
```

```
# Now remove the duplicated values.
df.drop_duplicates(inplace=True)
```

```
# Now, checking again for duplicate values.
df.duplicated().sum()
```

```
# Find duplicate values.
df.duplicated().sum()
```

**Find if the dataset contains any null values**

```
df.isnull().sum()
```

**Check empty columns**

```
df.info()
```

From the above data, it is clear that the 'Status' and 'unnamed' columns are empty, thus I'll delete them.

```
df.drop(['Status', 'unnamed1'], axis=1, inplace=True)
```

Again, checking the dataset.

```
df.info()
```

Since the empty columns are removed, now we'll check for null values.

```
df.isnull().sum()
```

```
df[df.isnull().any(axis=1)]
```

Since 12 records have null values, they should be dropped.

```
df.dropna(inplace=True)
```

```
df.isnull().sum()
```

```
df.info()
```

```
# to obtain mathematical description of different columns
df[['Age', 'Orders', 'Amount']].describe()
```

## Exploratory Data Analysis

```
df.columns
```

**Gender distribution of consumers**

```
ax = sns.countplot(x='Gender', data=df)

# To show values as well
for bars in ax.containers: 
    ax.bar_label(bars)
ax.set_ylabel('Number of consumers')
ax.set_title('Gender distribution')
```

**Finding the amount of sales done by each gender group**

```
s = df.groupby('Gender').sum()['Amount'].sort_values()
# To display data in proper tabulated format
display(HTML(s.to_frame().to_html()))
```

```
# Visualizing the amount of sales done by each gender.
plt.bar(s.index, s.values, color = ['tab:cyan', 'tab:purple'])
plt.xlabel('Gender')
plt.ylabel('Amount')
plt.title('Amount of sales done by each gender')
```

***Above visualization reveals that most consumers during the Diwali sales were female and their purchase was also greater than men.***

### Age

```
ax = sns.countplot(x='Age Group', data=df, hue = 'Gender')
for bars in ax.containers:
    ax.bar_label(bars)
ax.set_ylabel('Number of consumers')
ax.set_title('Age distribution')
```

```
# Distribution of sales amount as per age group
sales_age = df.groupby('Age Group').sum()['Amount']
plt.figure(figsize=(15,5))
sns.barplot(x = sales_age.index, y = sales_age.values)
```

***From above graph, it is clear that the individuals in the 26-35 age group contributed to the highest number of sales***

### State

```
#Total number of orders from top 10 states
sales_state =df.groupby('State')['Orders'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(20,5))
sns.barplot(x = sales_state.index, y = sales_state.values)
plt.ylabel('Number of orders')
plt.title('Total orders of top 10 states')
plt.xticks(rotation=90)
plt.show()
```

```
# Total amount of sales from top 10 states
sales_amt_state = df.groupby('State')['Amount'].sum().sort_values(ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)}) #Set graph size
sns.barplot(x = sales_amt_state.index, y = sales_amt_state.values)
plt.ylabel('Total sales')
plt.title('Total sales of top 10 states')
plt.xticks(rotation=90)
plt.show()
```

***Above graphs show that the top 10 total sales are commensurate to the top 10 orders and the most sales are credited to UP, Maharashtra and Karnataka.***

### Marital Status

```
# Total number of consumers according to their marital status
ax = sns.countplot(data = df, x = 'Marital_Status', palette = [sns.color_palette()[-1], sns.color_palette()[4]])
for bars in ax.containers:
    ax.bar_label(bars)
plt.ylabel('Number of consumers')
plt.title('Total number of consumers according to their marital status')
plt.show()
```

```
# Marital status wrt gender
sales_marital = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by = 'Amount', ascending=False)

sns.barplot(data = sales_marital, x = 'Marital_Status', y = 'Amount', hue = 'Gender', palette = [sns.color_palette()[-1], sns.color_palette()[4]])
plt.ylabel('Total sales')
plt.title('Total sales with respect to marital status and gender')
plt.xticks(rotation=90)
plt.show()
```

***From the graphs, it is clear that the most sales are done by unmarried women.***

### Occupation

```
# Total number of consumers according to their occupation
sns.set(rc={'figure.figsize':(20,5)}) #Set graph size
ax = sns.countplot(data = df, x = 'Occupation')
for bars in ax.containers:
    ax.bar_label(bars)
plt.ylabel('Number of consumers')
plt.title('Total number of consumers according to their occupation')
plt.xticks(rotation=90)
plt.show()
```

```
# Sales amount according to occupation
sales_amt_occupation = df.groupby('Occupation')['Amount'].sum().sort_values(ascending=False)
sns.set(rc={'figure.figsize':(20,5)}) # Set graph size
sns.barplot(x = sales_amt_occupation.index, y = sales_amt_occupation.values)
plt.ylabel('Total sales')
plt.title('Total sales according to occupation')
plt.xticks(rotation=90)
plt.show()
```

### Product Category

```
# Total number of products according to their category
sns.set(rc={'figure.figsize':(20,5)}) #Set graph size
ax = sns.countplot(data = df, x = 'Product_Category')
for bars in ax.containers:
    ax.bar_label(bars)
plt.ylabel('Number of products')
plt.xlabel('Product Category')
plt.title('Total number of products according to their category')
plt.xticks(rotation=90)
plt.show()
```

```
# Sale of each product
sales_amt_occupation = df.groupby('Product_Category')['Amount'].sum().sort_values(ascending=False)
sns.set(rc={'figure.figsize':(20,5)}) # Set graph size
sns.barplot(x = sales_amt_occupation.index, y = sales_amt_occupation.values)
plt.ylabel('Total sales')
plt.xlabel('Product Category')
plt.title('Sale amount of each product')
plt.xticks(rotation=90)
plt.show()
```

***From above data, it is clear that Food, Clothing, electronics, and Shoes contribute to the most sales***

## Conclusion

***From the above analysis, it is clear that unmarried women between the ages of 26-35 years from UP, Maharastra and Karnataka involved in IT sector, Healthcare and Aviation are more likely to buy Food, Clothing and Electronic products***
